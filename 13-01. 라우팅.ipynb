## 01. 라우팅 방식의 이해
- 웹 애플리케이션에서 라우팅 (Routing)이란 사용자가 주소 (URL)를 입력하거나 버튼을 클릭했을 때, 그에 맞는 화면(UI)를 보여주는 기능
- 웹 애플리케이션에의 구조는 크게 SPA와 MPA로 구분되며, 이 구조는 렌더링 방식과도 관련

<br>

### SPA와 CSR
- 싱글 페이지 애플리케이션 (Single Page Application, SPA)는 하나의 HTML 페이지로 구성된 웹 애플리케이션
- 사용자가 웹 사이트에 접속할 때 전체 애플리케이션을 한 번에 불러오며, 이후에는 URL이 변경되더라도 페이지 전체를 다시 로드하지 않고 필요한 부분만 화면에 다시 그려줌

  
<br>

#### 클라이언트 사이드 렌더링 (CSR, Client Side Rendering)
- 화면을 서버가 아닌 사용자의 웹 브라우저에서 자바스크립트를 이용해 직접 렌더링 하는 방식
  
  $\rightarrow$ 이때 서버는 초기 HTML만 전달하고, 이후 필요한 데이터는 API를 통해 서버에서 받아옴
  
  $\rightarrow$ 웹 브라우저는 이 데이터를 바탕으로 UI를 구성

  - React와 같은 SPA 프레임워크에서는 클라이언트 사이드 렌더링 방식을 주로 이용

- SPA 구조는 URL이 바뀌더라도 전체 페이지를 새로 불러오지 않고, 필요한 컴포넌트만 업데이트할 수 있어 CSR과 잘 어울림

<img src='img/1301.png' width=400>

<br>

- SPA와 CSR 조합은 한 번 페이지가 로드되면 이후 화면 전환이 빠르고 부드러워 사용자 경험이 우수
- 필요할 떄만 데이터를 받아오기 때문에 효율적
- **하지만 초기 로딩 시 자바스크립트 번들까지 함꼐 받아야 하므로, 첫 화면이 늦게 뜰 수 있으며, 콘텐츠가 웹 브라우저에서 동적으로 생성되어 검색 엔진 최적화에 불리함**

<br>

### MPA와 SSR
- 멀티 페이지 애플리케이션 (Multiple Page Application)은 여러 HTML 페이지로 구성된 애플리케이션
- MPA 구조에서는 사용자가 다른 페이지로 이동하면 웹 브라우저는 해당 경로에 맞는 새로운 HTML 파일을 서버에 요청하고, 전체 페이지를 다시 불러옴

<br>

#### 서버 사이드 렌더링 (SSR, Server Side Rendering)
- 서버에서 HTML을 미리 렌더링해 클라이언트에 완성된 페이지를 전달
  
  $\rightarrow$ 이떄 서버는 요청된 경로에 따라 HTML을 직접 렌더링하거나 미리 만들어진 HTML 파일을 찾아 클라이언트로 전송

- MPA는 여러 페이지로 구성되어있어, SSR과 자연스럽게 연결됨
  - 요청된 URL에 따라 서버는 그에 맞는 HTML을 생성하거나 찾아서 응답

<img src='img/1302.png' width=400>

<br>

- **MPA와 SSR 조합은 서버가 미리 렌더링한 HTML을 전달하므로 초기 로딩 속도가 빠르고, 콘텐츠가 HTML에 포함되어있어 SEO에 유리**
- **하지만, 페이지를 ㅣ동할 때마다 서버에서 새로운 HTML을 받아와야 해서 반응 속도가 느리고, 전체 페이지를 매번 새로 불러옴**

<br>

### React와 라우터
- React는 전통적인 MPA와 달리 SPA 구조의 웹 애플리케이션을 만드는 프런트엔드 라이브러리
- 이런 SPA 환경에서는 페이지를 이동할 때마다 서버에 새로운  HTML을 요청하는 대신, 클라이언트에서 URL을 동적으로 처리해야 함
  
  $\rightarrow$ 웹 브라우저가 제공하는 히스토리 API를 활용 (History API)

  $\rightarrow$ 자바스크립트로 URL을 변경하거나 이전/다음 페이지로 이동하는 동작을 제어
- URL에 따라 어떤 컴포넌트를 렌더링할지 결정하거나 화면 레이아웃을 동적으로 변경하려면, 많은 조건 분기와 코드가 필요
  
  $\rightarrow$ **React Router는 React 애플리케이션에서 CSR 기반 라우팅을 간편하게 구현할 수 있도록 도와주는 라이브러리**

  - **웹 브라우저의 히스토리 API를 내부적으로 활용해 URL이 변경되면 해당 경로에 맞는 컴포넌트를 렌더링해줌**
  
<br>

#### React Router의 기능
1. **URL에 따른 컴포넌트 렌더링 : 경로에 따라 해당 컴포넌트를 자동으로 표시**
- 예: `/home` $\rightarrow$ Home 컴포넌트
  
2. **중첩 라우팅 : 한 페이지 안에 여러 컴포넌트를 중첩해 하위 경로별로 다른 화면을 구성 가능**
- 예: Dashboard 컴포넌트 내부에 여러 하위 경로 구성 (`/dashboarde/profile` 등)

3. **동적 라우팅 : URL에 변수처럼 동적인 값을 넣어 특정 데이터를 표시 가능**
- 예: `/user/:id` $\rightarrow$ UserDetail 컴포넌트에서 해당 사용자의 정보 로드

4. **프로그래밍 방식 내비게이션 : 사용자가 직접 링크를 클릭하지 않아도, 자바스크립트 코드로 페이지 전환을 제어 가능**
- 로그인 후 자동으로 메인 페이지 이동   

5. **리다이렉션 및 조건부 렌더링 : 특정 조건에 따라 다른 경로로 이동시키거나 조건에 맞는 컴포넌트만 보여줌**
- 로그인핮 않은 사용자는 `/login` 페이지로 강제 이동

  $\rightarrow$ 각 경로에 연결할 컴포넌트를 따로 만들어야 함 (`/pages` 디렉토리)

### 라우팅 범위 지정
- **React Router를 기존 React 애플리케이션에 적용하려면 가장 먼저 라우팅 컨텍스트를 지정해야 함**
- **라우팅 컨텍스트 : 라우팅 기능이 적용될 범위를 설정하는 작업**
  - Context API에서 전역 상태를 사용할 범위를 지정했던 것과 유사

<br>

- **React Router에서는 용도에 따라 다음과 같이 4가지 라우팅 컨텍스트 컴포넌트를 제공**
  - **`BrowserRouter` : 최신 웹 브라우저에서 제공하는 히스토리 API를 사용해 URL을 관리**
    - 가장 일반적이며, CSR 환경에서 주로 사용
    - 서버가 아닌 클라이언트에서 URL을 처리하므로 현대 웹 애플리케이션에서 가장 적합
  - **`HashRouter` : URL에 해시(`#`) 기호를 포함해 경로를 구분**
    - 서버 설정이 필요 없기 때문에, 정적 파일 기반 환경에서 유용
  - **`MemoryRouter` : 주소창을 변경하지 않고, 내부 메모리에서 라우팅 상태를 관리**
    - 주로 테스트 환경이나 MPA 구조에서 사용
    - 변경한 URL을 표시하지 않기 떄문에, 웹 브라우저의 히스토리나 주소창에 영향을 미치지 않음
  - **`StaticRouter` : SSR 환경에서 사용하는 라우터**
    - 클라이언트의 주소창은 변경하지 않고 서버에서 HTML을 렌더링할 떄 사용

<br>

- **실무에서 가장 널리 사용하는 컴포넌트는 `BrowserRouter`**
  
  $\rightarrow$ 현대 웹 브라우저에서 지원하는 URL 구조와 가장 잘 맞고, CSR 환경에서 라우팅을 구현하기에 적합하기 때문
- `main.tsx`

```ts
import { StrictMode } from "react";
import { createRoot } from "react-dom/client";
import { BrowserRouter } from "react-router";
import App from "./App.tsx";


createRoot(document.getElementById("root")!).render(
  <StrictMode>
    <BrowserRouter>
      <App />
    </BrowserRouter>
  </StrictMode>
);
```

<br>

### 라우트 설정
- **`BrowserRouter` 컴포넌트로 라우팅 범위를 지정하고 난 후, 라우트 설정**
- **라우트 설정 (Configuring Route)은 특정 URL 경로를 컴포넌트와 연결하는 작업**
  
  $\rightarrow$ 사용자가 웹 사이트의 특정 주소로 방문했을 때, 어떤 컴포넌트를 보여줄지 결정하는 것

  - **`Route` : 하나의 라우트 정보를 정의. `path` 속성에 URL 경로를, `element` 속성에 해당 경로에 표시할 컴포넌트를 지정**
  - **`Routes` : 여러 개의 `Route` 컴포넌트를 감싸는 컨테이너 역할. 내부에서 경로를 비교해, 현재 URL과 일치하는 첫 번째 `Route` 컴포넌트를 렌더링**  

<br>

- **`Route` 컴포넌트의 주요 속성**
    - `element` 속성은 필수이며, 나머지는 선택사항

| **속성** | **타입** | **설명** |
| - | - | - |
| `path` | `string` | URL 경로 지정|
| `element` | `React.ReactNode` | 해당 경로에서 렌더링할 컴포넌트 |
| `index` | `boolean` | 부모 라우트의 기본 페이지 지정 여부 |
| `children` | `React.ReactNode` | 중첩 라우틀 정의할 때 사용 |
| `errorElement` | `React.ReactNode` | 해당 라우트에서 오류 발생 시 표시할 컴포넌트 |
| `handle` | `object` | 라우트에 메타 데이터를 전달할 때 사용 |
| `lazy` | `function` | 동적 임포트를 통해 코드 분할 구현  |


<br>

- `App.tsx`
  - `/` 경로에는 `Home` 컴포넌트를
  - `/about` 경로에는 `About` 컴포넌트를 연결하는 라우트 설정 예제

```ts
import { Route, Routes } from "react-router";
import Home from "./pages/Home";
import About from "./pages/About";

export default function App() {
  return (
    <Routes>
      <Route path="/" element={<Home />} />
      <Route path="about" element={<About />} />
    </Routes>
  );
}
```

## 03. React Router 기능
- React Router는 더 복잡한 UI 구성과 경로 처리를 위해 다양한 기능을 제공

<br>

### 중첩 라우트
- **중첩 라우트 (Nested Rout)란 하나의 라우트 안에 또 다른 라우트를 포함하는 방식**
  - 복잡한 UI를 계층적으로 관리하고, 하위 페이지 구조도 효율적으로 구성 가능

<br>

- `src/pages/Dashboard.tsx`
  - **`Outlet` 컴포넌트 : 하위 라우트가 렌더링 될 자리를 표시**

```ts
import { Outlet } from "react-router";

export default function Dashboard() {
  return (
    <> 
      <h1>Dashboard</h1>
      <Outlet />
    </>
  )
}
```

<br>

- `src/pages/Summary.tsx`

```ts
export default function Summary() {
    return <h1>Summary</h1>;
}
```

<br>

- `src/pages/Settings.tsx`

```ts
export default function Settings() {
    return <h1>Settings</h1>;
}
```

<br>

- `App.tsx`
    - **부모 컴포넌트를 기준으로 `index` 속성이 붙은 라우트는 하나만 지정 가능**
    - **`index` 속성을 지정한 컴포넌트는 하위 라우트를 가질 수 없음 $\rightarrow$ 다른 중첩 라우트의 부모가 될 수 없음**
  
```ts
import { Route, Routes } from "react-router";
import Home from "./pages/Home";
import About from "./pages/About";
import Dashboard from "./pages/Dashboard";
import Settings from "./pages/Settings";
import Summary from "./pages/Summary";

export default function App() {
  return (
    <Routes>
      <Route path="/" element={<Home />} />
      <Route path="about" element={<About />} />
      <Route path="dashboard" element={<Dashboard />}>
        <Route index element={<Summary />} />
        <Route path="settings" element={<Settings />} />
      </Route>
    </Routes>
  );
}
```

### 레이아웃 라우트 (Layout Route)
- **React Router는 여러 페이지에서 공통으로 사용하는 UI 요소 (헤더, 푸터 등)를 효율적으로 관리할 수 있는 기능도 제공**
  
  $\rightarrow$ **레이아웃 라우트**

- **레이아웃 라우트 : 여러 페이지에서 공통으로 사용하는 레이아웃을 정의하는 라우트**
  - **중첩 라우트는 `path` 속성으로 특정 경로에 대응하지만, 레이아웃 라우트는 `path` 속성을 사용하지 않고, 자식 라우트를 감싸는 데만 사용**

    $\rightarrow$ **부모 라우트에 `path` 속성이 잇다면 중첩 라우트, 없으면 레이아웃 라우트로 작동**

<br>

- `src/layout/RootLayout.tsx`
  - **레이아웃 라우트에서는 `Outlet` 컴포넌트를 반드시 포함해야 함**

```ts
import { Outlet } from "react-router";

export default function RootLayout() {
  return (
    <>
        <header>
            <h1>My Application</h1>
        </header>
        <main>
            <Outlet />
        </main>
        <footer>
            <p>© 2024 My Application</p>
        </footer>
    </>
  );
}
```

<br>

- **레이아웃 라우트 적용**
- `App.tsx`

```ts
import RootLayout from "./layouts/RootLayout";

export default function App() {
  return (
    <Routes>
      <Route path="/" element={<RootLayout />}>
        <Route path="/" element={<Home />} />
        <Route path="about" element={<About />} />
        <Route path="dashboard" element={<Dashboard />}>
          <Route index element={<Summary />} />
          <Route path="settings" element={<Settings />} />
        </Route>      
      </Route>
    </Routes>
  );
}
```

### 라우트 프리픽스 (Route Prefix)
- 특정 그룹의 라우트 경로에 공통된 접두사 (Prefix)를 붙임

<br>

- `/dashboard`를 포함하는 중첩 라우트 그룹에 `/my`라는 접두사 추가
- `App.tsx`

```ts
export default function App() {
  return (
    <Routes>
      <Route path="/" element={<RootLayout />}>
        <Route path="/" element={<Home />} />
        <Route path="about" element={<About />} />
        <Route path='my'>
          <Route path="dashboard" element={<Dashboard />}>
            <Route index element={<Summary />} />
            <Route path="settings" element={<Settings />} />
          </Route>
        </Route>
      </Route>
    </Routes>
  );
}
```

<br>

### 동적 세그먼트 (Dynamic Segment)
- **React Router에서는 URL 경로 중 일부를 동적으로 바꿀 수 있는 값으로 지정 $\rightarrow$ 동적 세그먼트**
- **Route 컴포넌트의 `path` 속성에 콜론 (`:`)을 붙이면 해당 부분이 동적 세그먼트로 처리**

<br>

- `App.tsx`
  - `path='team/:teamId` $\rightarrow$ `/team/1`, `/team/2` ... 등 다양한 ID 경로 연결 가능

```ts
import Team from "./pages/Team";

export default function App() {
  return (
    <Routes>
      <Route path="/" element={<RootLayout />}>
        <Route path="team/:teamId" element={<Team />} />
        
      </Route>
    </Routes>
  );
}
```

<br>

- **URL에 포함된 동적 세그먼트 값을 컴포넌트에서 사용하려면 React Router에서 제공하는 `useParams` 훅을 사용**

<br>

- `src/pages/Team.tsx`
  - `params.teamId`는 URL에서 `:teamId`에 대응하는 실제 값
    - 예) `/team/3` $\rightarrow$ `params.teamId` 는 3

```ts
import { useParams } from "react-router";

export default function Team() {
    const params = useParams();
    return (
        <>
            <h1>Team {params.teamId}</h1>
        </>
    )
}
```

### 옵셔널 세그먼트 (Optional Segment)
- **React Router에서는 URL 경로의 일부를 선택해 입력할 수 있도록 처리 $\rightarrow$ 옵셔널 세그먼트**
  
  세그먼트 뒤에 물음표 (`?`)를 붙여 설정

- 옵셔널 세그먼트를 사용하면 일부 세그먼트가 없어도 해당 라우트가 작동하도록 만들 수 있음
- 다음은 `/team/1/group/1`, `/group/1`경로 모두에 매칭 $\rightarrow$ 즉 `team`세그먼트와 동적 세그먼트인 `teamId`도 생략 가능


```ts
<Route path='team?/:teamId?/group/:groupId' element={<Team />}>
```

- **단, `useParams()` 훅을 사용할 때 옵셔널 세그먼트는 URL에 실제값이 포함되어야만 `params` 객체 값이 들어옴**
  
  **즉, 세그먼트가 생략되었다면 해당 값은 `undefined`로 반환 $\rightarrow$ 옵셔널 세그먼트를 사용하는 라우트에서는 `useParams` 훅으로 받은 값이 무조건 존재한다고 가정하면 안됨**

- **이미 `/` 경로에 해당하는 `Route`가 있다면 라우트 충돌이 발생할 수 있으며, 경로를 명확하게 정의하기 어려워지므로 사용하지 않는 것이 권장**

### 스플랫 (Splat)
- **정의되지 않은 라우트를 포괄적으로 처리하는 기능 (Catch-All 또는 와일드카드)**
- **경로에 `*` 기호를 사용해 설정 $\rightarrow$ 특정 경로 이후의 모든 하위 경로를 한꺼번에 매칭 가능**
- `Route path='*'`로 설정한 스플랫 라우트는 작성 위치에 따라 작동 방식이 달라짐
  - 레이아웃 라우트 내부에 위치 $\rightarrow$ 공통 레이아웃의 `Outlet` 위치에
  - 레이아웃 라우트 외부에 위치 $\rightarrow$ 공통 레이아웃 없이 `NotFound` 컴포넌트만 단독으로 렌더링

<br>

- **스플랫은 존재하지 않는 경로 (404 Not Found)를 처리할 때 자주 사용**
- `src/pages/NotFound.tsx`

```ts
export default function NotFound() {
    return <h1>404 - Page Not Found</h1>;
}
```

<br>

- `App.tsx`

```ts
export default function App() {
  return (
    <Routes>
      <Route path="/" element={<RootLayout />}>

        ...

      <Route path="*" element={<NotFound />} />
    </Routes>
  );
}
```

### 문서 메타데이터 설정 (React 19 이후)
- React 19이후 HTML의 `<title>`, `<link>`, `<meta>`, `<script async>` 등의 메타데이터 태그를 JSX 안에서 직접 사용할 수 있는 기능이 추가

  이전까지는 별도의 외부 라이브러리 (`react-helmet`)을 사용하거나 `index.html`을 수정해야 했던 작업을 이제는 각 컴포넌트 내부에서 직접 JSX형태로 작성하면 자동으로 `<head>`에 반영

<br>

- `src/pages/Home.tsx`

```ts
export default function Home() {
  return (
    <>
      <h1>Home</h1>
      <title>React 19</title>
      <link rel='author' href='https://dmc-datahome-996518370991.asia-northeast3.run.app/dcdp/' />
      <meta name='keywords' content='React, Vite, JavaScript' />
      <script async src='https://example.com/some-script.js'></script>    
    </>
  )
}
```

#### `<title>` 태그
- 단순한 정적 문자열뿐만 아니라 동적 세그먼트나 `props` 값 등 변화하는 데이터를 기반으로 구성 가능
  - 예) URL 경로에 포함된 `teamId`와 `groupId`를 기반으로 제목 설정 가능
  - **JSX표현식 (`{}`)를 섞어 사용하는 방식은 허용되지 않음**
- **`<title>` 태그를 여러 컴포넌트에서 중복으로 작성할 수는 있지만, 웹 브라우저는 가장 마지막에 렌더링된 하나만 적용**
  
  $\rightarrow$ **하나의 페이지 안에 `<title>` 태그가 중복되지 않도록 구성해야 하며, 중첩 라우트나 레이아웃 컴포넌트를 사용할 때 주의 필요**

<br>

- `src/pages/Team.tsx`

```ts
import { useParams } from "react-router";

export default function Team() {
    const params = useParams();
    return (
        <>
            <h1>Team {params.teamId} | Group {params.groupId}</h1>
            <title>
                {`Team ${params.teamId} | Group ${params.groupId}`}
            </title>
        </>
    )
}
```
