## 01. JSX (Javascript XML)
- **자바스크립트 코드 안에서 HTML과 비슷한 문법을 사용해 UI를 정의할 수 있도록 도와주는 문법 확장 (Syntatic Extention)**
  - 자바스크립트 안에서 HTML처럼 생긴 코드를 쓸 수 있게 해주는 문법
- JSX는 실제로는 HTML이 아니며, React가 JSX코드를 JS코드로 변환해서 실행
    - React는 `create Element()`호출 코드로 바꾸어 실행
    - React에서 JSX는 필수가 아님

```js
<div>
    <h1>Hello, React!</h1>
</div>
```

## 02. JSX 문법

<br>

### 하나의 루트 요소로 반환
- **JSX로 작성된 컴포넌트는 반드시 하나의 태그(요소)만 반환해야 함**
- **프래그먼트 (Fragment) : 여러 요소를 묶어주는 가상의 태그**
  - 실제 HTML에는 아무것도 렌더링 되지 않기에 불필요한 태그가 추가되지 않음

```js
// <React.Fragment> 혹은 <>
<React.Fragment> 
    <div>첫 번째 영역</div>
    <div>두 번째 영역</div>
</React.Fragment>
```

<br>

### 모든 태그 닫기
- HTML에서는 일부 태그를 닫지 않아도 작동 (self-closing tag)
  - `<br>`, `<img>`, `<input>` 등
- **JSX에서는 빈 태그라도 자체 종료 태그를 사용해야 함**
  
```js
<br>
<br />
```

- **JSX 변환 도구 : https://transform.tools/html-to-jsx**

<br>

### 태그 속성은 카멜 케이스로 작성
- **카멜 케이스 (camelCase) : 첫 번째 단어는 첫 글자를 소문자, 두 번째 단어부터는 첫 글자를 대문자로 표기**
- JSX는 자바스크립트 문법과 함꼐 동작
  
  JSX에서는 속성명을 JS 변수명처럼 카멜 케이스로 작성

- HTML
  
```html
<div class="box"></div>
```

- JSX

```js
<div className="box"></div>
```

<br>

- **HTML과 JSX에서 서로 이름이 다른 속성**
  - `data-`나 `aria-`로 시작하는 속성은 HTML 표준 규칙을 따름

| HTML 속성명 | JSX 속성명 |
|-----------------|---------------------|
| class           | className           |
| for             | htmlFor             |
| contenteditable | contentEditable     |
| tabindex        | tabIndex            |
| readonly        | readOnly            |
| maxlength       | maxLength           |
| colspan         | colSpan             |
| rowspan         | rowSpan             |
| frameborder     | frameBorder         |
| usemap          | useMap              |
| accesskey       | accessKey           |
| enctype         | encType             |
| accept-charset  | acceptCharset       |
| inputmode       | inputMode           |
| autoplay        | autoPlay            |
| formnovalidate  | formNoValidate      |

<br>

### 표현식 중괄호 사용
- **JSX에는 중괄호 (`{}`) 안에 JS 표현식을 넣어 사용 가능 $\rightarrow$ 화면에 표시할 내용을 동적으로 개발 가능**

```js
<h1>result: {10 + 20}</h1>
```

<br>

- JSX안에서 삼항 연산 사용 가능

```js
<h1>{10 <ㅗ 20 ? "10이 더 작다" : "20이 더 크다"}</h1>
```

<br>

- **논리 연산자 (&&) 사용할 때, 조건이 참일 때만 JSX 요소를 출력**

```js
<>
  {10 > 20 && <h1> 10이 20보다 더 크다</h1>}
  {10 < 20 && <h1> 10이 20보다 더 작다</h1>}
</>
```

<br>

### 인라인 스타일 객체 지정
- HTML에서는 태그에 직접 `style` 속성을 지정 $\rightarrow$ 인라인 스타일

```html
<h1> style="color: red; background-color: lightgray">Stylde Text<h1>
```

<br>

- **JSX에서는 `style` 속성에 JS객체를 할당**
  
  **CSS 속성은 반드시 카멜 케이스로 작성 ('-' 포함시 카멜케이스로)**

```js
<h1
  style={{
    color: "red",
    backgroundColor: "lightgray",
  }}
>Styled Text
</h1>
```

<br>

### 중괄호 내부 주석

```js
<>
  {/* 주석 */}
  <div>주석</div>
</>
```
